Here we are all set to run excat same code for all Days of the week

```python
ss = openpyxl.load_workbook(file_path)
sheetnames_list = ss.sheetnames
```

```python
Monday_index = sheetnames_list.index("MONDAY")
Days = sheetnames_list[Monday_index:]
Days
```

```python
time = []
for day in Days:
    df = pd.read_excel(file_path, sheet_name=day, header=2)
    # display(HTML(df.to_html()))
    curr_day_time = list(df.columns)[1:]
    # time = time + curr_day_time
    # time  = sorted(list(set(time)) + list(set(curr_day_time)))
    print(curr_day_time)
    time  = time + [curr_time for curr_time in curr_day_time if curr_time not in time]
    
print("\n\n")
print(time)
```

### PUTTING IT ALL TOGETHER FOR EACH DAYS

```python
def func(series):
    return (series.apply(lambda x: my_section in str(x) and exclude_combine_section_course not in str(x)))

timeTable = pd.DataFrame()
timeTable["Time"] = time
timeTable.set_index("Time", inplace=True)

for day in Days:
    df = pd.read_excel(file_path, sheet_name=day, header=2)
    
    # Transposing the data-frame so that index becomes the time of the classes/
    df = df.T
    
    #Removing the Serial number(after transposing) as column names   #Don't run it more than once          
    df.columns = df.iloc[0]
    df = df.drop(df.index[0])
    df = df.drop(columns= "CLASSROOMS")   # unneccessary column

    #replacing "\n" with ","
    for col in df.columns:
        df[col] = df[col].apply(lambda x: str(x).replace("\n"," , "))

    mask  = df.apply(lambda x:  func(x), axis=1)
    # print(mask)  
    classes = df[mask]
    classes = classes.dropna(how='all', axis=0)
    classes = classes.dropna(how='all', axis=1)
    classes = classes.replace(np.nan, "-")
    
    # Below code will squeeze the multiple columns into single one

    # concatenate the Column names with its values i.e Concatenating Room numbers e.g (R12)
    for col in classes.columns:
        classes[col] = classes[col].apply(lambda x: str(x) + "," + col)
    # replacing dash + column name values with empty strings  
    for col in classes.columns:
        classes[col] = classes[col].apply(lambda x: str(x).replace( "-,"+ col, ""))

    # merging all columns into a single Column
    cols = classes.columns.to_list()
    timeTable[day] = classes[cols].agg("".join, axis=1)
```

```python
TTable = timeTable.copy()
TTable.replace(np.nan, "-", inplace=True)
TTable.to_csv("TimeTable.csv", index=True)
```

```python
def func2(input1):
    if any( course in str(input1) and "BSE-6A" in str(input1) for course in ElectiveCourses ):
        return True
    elif (my_section in str(input1) and exclude_combine_section_course not in str(input1)):
         return True
    elif ("BCS-8A" in input1 and "DLP" in input1):
         return True
    else : return False

def func(series):
    return (series.apply(func2))


timeTable = pd.DataFrame()
timeTable["Time"] = time
timeTable.set_index("Time", inplace=True)

for day in Days:
    df = pd.read_excel(file_path, sheet_name=day, header=2)
    
    # Transposing the data-frame so that index becomes the time of the classes/
    df = df.T
    
    #Removing the Serial number(after transposing) as column names   #Don't run it more than once          
    df.columns = df.iloc[0]
    df = df.drop(df.index[0])
    df = df.drop(columns= "CLASSROOMS")   # unneccessary column

    #replacing "\n" with ","
    for col in df.columns:
        df[col] = df[col].apply(lambda x: str(x).replace("\n"," , "))

    mask  = df.apply(lambda x:  func(x), axis=1)
#     print(mask)
    
    classes = df[mask]
    classes = classes.dropna(how='all', axis=0)
    classes = classes.dropna(how='all', axis=1)
    classes = classes.replace(np.nan, "-")
    
    # Below code will squeeze the multiple columns into single one

    # concatenate the Column names with its values i.e Concatenating Room numbers e.g (R12)
    for col in classes.columns:
        classes[col] = classes[col].apply(lambda x: str(x) + "," + col)
    # replacing dash + column name values with empty strings  
    for col in classes.columns:
        classes[col] = classes[col].apply(lambda x: str(x).replace( "-,"+ col, ""))

    # merging all columns into a single Column
    cols = classes.columns.to_list()
    timeTable[day] = classes[cols].agg("".join, axis=1)
```

```python
TTable = timeTable.copy()
TTable.replace(np.nan, "-", inplace=True)
TTable.to_csv("TimeTableElective.csv", index=True)
display(HTML(TTable.to_html()))
```
