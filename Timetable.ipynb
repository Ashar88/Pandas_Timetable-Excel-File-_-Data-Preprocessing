## Timetable Excel File - Preprocessing

#### Including Required Libraries

```python
import pandas as pd
import numpy as np
```

```python
file_path = "FAST School of Computing - Fall 2022 - V1.xlsx"
my_section = "BSE-5A"
exclude_combine_section_course = "BAI";
sheet_ = "BS(SE) BATCH 2020"
```

### Reading Courses and Faculty name for my FALL-22 Semester

```python
df = pd.read_excel(file_path , sheet_name= sheet_, header=4,
                   skiprows=lambda  x: x in [5], skipfooter=1, index_col=2)
#Skipping invalid rows
```

```python
df.head(5)
```

#### Removing the Redundant column

```python
df = df.drop(columns='Unnamed: 0')  
df
```

#### Removing  Rows having most values NULL

```python
def detectNullRow(x):
    # print(pd.isna(x.name))
    if(~pd.isna(x.name)):      #x.name should be a valid obj/str
        return(df.loc[x.name].isnull().sum() >= 3)   #if null value count exceed 3, then that row needs to be removed
    
mask = df.apply(lambda x: detectNullRow(x) , axis=1)
indexes = df.loc[mask].index

df = df.drop(index = indexes)
```

```python
df
```

#### Replacing "\n" with commas (optional)

```python
df["Instructors (Sections)"] = df["Instructors (Sections)"]. apply(lambda x: x.replace("\n", " , "))
```

### Now Reading the Daywise Timetable of my section

First, we try to get Timetable for Monday only, afterwards, we loop it over for other Days 

```python
df = pd.read_excel(file_path, sheet_name="MONDAY", header=2)
```

```python
df.head(3)
```

```python
# Transposing the data-frame so that index would be the time of the classes(better in terms of understanding the data)
df = df.T   
df.head(3)
```

```python
#Removing the column indexes (after transposing) as they are of no meaning.    #Don't run it more than once  
#Making classroom names as column indexes now
df.columns = df.iloc[0]
df = df.drop(df.index[0]) 
df.head(3)
```

```python
df = df.drop(columns= "CLASSROOMS")   # unneccessary column
df.head(2)
```

#### Replacing "\n" with commas for all columns (optional)

```python

for col in df.columns:
    df[col] = df[col].apply(lambda x: str(x).replace("\n"," , "))
df.head(2)
```

```python
# saving the timing slots
time = df.index.to_list()
time
```

#### Finding My Section and carefully Excluding combine-section Courses

```python
def func(series):
     return (series.apply(lambda x: my_section in str(x) and exclude_combine_section_course not in str(x)))

    
mask  = df.apply(lambda x:  func(x), axis=1)
classes = df[mask]
```

#### Removing unnecessary Time-slots and Classrooms

```python
classes = classes.dropna(how='all', axis=0)
classes = classes.dropna(how='all', axis=1)
classes
```

Here we are all set to run excat same code for all Days of the week

### PUTTING IT ALL TOGETHER FOR EACH DAYS

```python
def func(series):
    return (series.apply(lambda x: my_section in str(x) and exclude_combine_section_course not in str(x)))

Days = ["MONDAY", "TUESDAY", "WEDNESDAY", "THURSDAY","FRIDAY"]
timeTable = pd.DataFrame()
timeTable["Time"] = time
timeTable.set_index("Time", inplace=True)

for day in Days:
    df = pd.read_excel("FAST School of Computing - Fall 2022 - V1.xlsx", sheet_name=day, header=2)
    
    # Transposing the data-frame so that index becomes the time of the classes/
    df = df.T
    
    #Removing the Serial number(after transposing) as column names   #Don't run it more than once          
    df.columns = df.iloc[0]
    df = df.drop(df.index[0])
    df = df.drop(columns= "CLASSROOMS")   # unneccessary column

    #replacing "\n" with ","
    for col in df.columns:
        df[col] = df[col].apply(lambda x: str(x).replace("\n"," , "))

    mask  = df.apply(lambda x:  func(x), axis=1)
    
    classes = df[mask]
    classes = classes.dropna(how='all', axis=0)
    classes = classes.dropna(how='all', axis=1)
    classes = classes.replace(np.nan, "-")
    
    # Below code will squeeze the multiple columns into single one

    # concatenate the Column names with its values i.e Concatenating Room numbers e.g (R12)
    for col in classes.columns:
        classes[col] = classes[col].apply(lambda x: str(x) + "," + col)
    # replacing dash + column name values with empty strings  
    for col in classes.columns:
        classes[col] = classes[col].apply(lambda x: str(x).replace( "-,"+ col, ""))

    # merging all columns into a single Column
    cols = classes.columns.to_list()
    timeTable[day] = classes[cols].agg("".join, axis=1)
```

```python
TTable = timeTable.copy()
TTable.replace(np.nan, "-", inplace=True)
TTable.to_csv("TimeTable.csv", index=True)
```
